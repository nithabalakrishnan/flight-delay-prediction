```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("CleanData/flight_info_2018.csv")

df.head()
```

```python
df.count()
```

```python
cleaned_df = df.drop(columns =['CANCELLED', 
                               'CANCELLATION_CODE',
                               'DIVERTED','CARRIER_DELAY',
                               'WEATHER_DELAY',
                               'NAS_DELAY',
                               'SECURITY_DELAY',
                               'LATE_AIRCRAFT_DELAY'])
cleaned_df.head()
```

```python
cleaned_df.fillna('NaN',inplace = True)
```

```python
cleaned_df.count()
```

```python
cleaned_df.dtypes
```

```python
cleaned_df.nunique()
```

```python
dropped_df = cleaned_df.replace(['NaN'],np.nan)
dropped_df = dropped_df.dropna(how = 'any')
dropped_df.head(10)
```
