```python
data = pd.read_csv("CleanData/delayed_flights_2018.csv")
data

df = data.drop(['Unnamed: 0'], axis=1)

df = df.drop(['Unnamed: 0.1'], axis=1)

df.head()
```

```python
df[["DAY_OF_WEEK", "DEP_TIME",'ORIGIN_AIRPORT_ID', "OP_CARRIER_FL_NUM", "DEP_DELAY", "ARR_DELAY", "DEST_AIRPORT_ID"]] = df[["DAY_OF_WEEK", "DEP_TIME",'ORIGIN_AIRPORT_ID', "OP_CARRIER_FL_NUM", "DEP_DELAY", "ARR_DELAY", "DEST_AIRPORT_ID"]].fillna(value=0)

df.head()
```

```python
len(df)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
dfle = df
dfle.OP_UNIQUE_CARRIER = le.fit_transform(dfle.OP_UNIQUE_CARRIER)

dfle.head()
```

```python
#reshaping


X = dfle[["DAY_OF_WEEK", "DEP_TIME",'ORIGIN_AIRPORT_ID', "OP_CARRIER_FL_NUM", "ARR_DELAY", "DEST_AIRPORT_ID"]]
y =  dfle[["DEP_DELAY"]].values.reshape(-1, 1)
print(X.shape, y.shape)



```

```python
X = X.round(0).astype(int)
y = y.round(0).astype(int)
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression
from sklearn.datasets import make_regression
```

```python
# Split the data into training and testing

### BEGIN SOLUTION
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=59)

```

```python
X_test.head()
```

```python
y_test[0:10]
```

```python
# Create a LinearRegression model and fit it to  training data

from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)

# predictions
# PREDICT DEP_DELAY in minutes

abb_X = X_test[0:10]
predictions = model.predict(abb_X)


predictions = predictions.round(0).astype(int)
predictions
```

```python
### BEGIN SOLUTION
predictions = model.predict(X_test[0:100])
model.fit(X_train[0:100], y_train[0:100])
plt.scatter(model.predict(X_train[0:100]), model.predict(X_train[0:100]) - y_train[0:100], c="blue", label="Training Data")
plt.scatter(model.predict(X_test[0:100]), model.predict(X_test[0:100]) - y_test[0:100], c="orange", label="Testing Data")
plt.legend()
plt.hlines(y=0, xmin=y_test[0:100].min(), xmax=y_test[0:100].max())
plt.title("Residual Plot")
plt.show()
```

```python
# Ridge model
# Note: Use an alpha of .01 when creating the model for this activity
from sklearn.linear_model import Ridge

### BEGIN SOLUTION
ridge = Ridge(alpha=.01).fit(X_train, y_train)

predictionsR = ridge.predict(X_test)

predictionsR = predictionsR.round(0).astype(int)
# MSE = mean_squared_error(y_test_scaled, predictions)
# r2 = ridge.score(X_test_scaled, y_test_scaled)
# ### END SOLUTION
predictionsR
# print(f"MSE: {MSE}, R2: {r2}")
```

```python
from sklearn.metrics import mean_squared_error
# call it with test data
MSE = mean_squared_error(y_test, predictionsR)
r2 = ridge.score(X_test, y_test)
print(f"MSE: {MSE}, R2: {r2}")
```
